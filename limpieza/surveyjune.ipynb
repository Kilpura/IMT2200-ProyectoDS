```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
ruta = os.path.join("..", "data", "Remote Work Of Health Impact Survey June 2025.csv")
df_survey = pd.read_csv(ruta, delimiter = ",")
df_survey
```

```python

#INDUSTRIA
replace_industry = {
    "Professional Services": "Servicios Profesionales",
    "Education": "Educación",
    "Manufacturing": "Manufactura",
    "Customer Service": "Atención al Cliente",
    "Technology": "Tecnología",
    "Finance": "Finanzas",
    "Retail": "Comercio Minorista",
    "Healthcare": "Salud",
    "Marketing": "Marketing"
}

#GENERO
replace_gender = {
    "Female": "Femenino",
    "Male": "Masculino",
    "Non-binary": "No binario",
    "Prefer not to say": "Prefiero no responder"
}

# CONTINENTES
replace_continent = {
    "Asia": "Asia",
    "Africa": "África",
    "Europe": "Europa",
    "South America": "América del Sur",
    "Oceania": "Oceanía",
    "North America": "América del Norte"
}
#ROLES DE TRABAJOS
replace_roles = {
    "Data Analyst": "Analista de Datos",
    "Business Analyst": "Analista de Negocios",
    "DevOps Engineer": "Ingeniero/a DevOps",
    "IT Support": "Soporte TI",
    "Technical Writer": "Redactor/a Técnico/a",
    "Software Engineer": "Ingeniero/a de Software",
    "HR Manager": "Gerente de RR. HH.",
    "Project Manager": "Jefe/a de Proyecto",
    "Data Scientist": "Científico/a de Datos",
    "UX Designer": "Diseñador/a UX",
    "Social Media Manager": "Gestor/a de Redes Sociales",
    "Digital Marketing Specialist": "Especialista en Marketing Digital",
    "Account Manager": "Ejecutivo/a de Cuentas",
    "Research Scientist": "Investigador/a Científico/a",
    "Sales Representative": "Representante de Ventas",
    "Operations Manager": "Gerente de Operaciones",
    "Customer Service Manager": "Jefe/a de Atención al Cliente",
    "Product Manager": "Gerente de Producto",
    "Quality Assurance": "Especialista QA",
    "Executive Assistant": "Asistente Ejecutivo/a",
    "Financial Analyst": "Analista Financiero/a",
    "Marketing Specialist": "Especialista en Marketing",
    "Consultant": "Consultor/a",
    "Content Writer": "Redactor/a de Contenidos"
}

# MODO DE TRABAJO
replace_work_mode = {
    "Onsite": "Presencial",
    "Hybrid": "Híbrido",
    "Remote": "Remoto"
}

#SALUD MENTAL
replace_mental_health = {
    "Stress Disorder": "Trastorno por estrés",
    "ADHD": "TDAH (Trastorno por Déficit de Atención e Hiperactividad)",
    "Burnout": "Síndrome de burnout",
    "Anxiety": "Ansiedad",
    "PTSD": "TEPT (Trastorno de Estrés Postraumático)",
    "Depression": "Depresión"
}

#Desgaste profesional
replace_desgaste = {
    "High": "Alto",
    "Medium": "Medio",
    "Low": "Bajo"
}
#Salud Fisica
replace_physical_health = {
    "Shoulder Pain": "Dolor de hombro",
    "Neck Pain": "Dolor de cuello",
    "Back Pain": "Dolor de espalda",
    "Eye Strain": "Fatiga visual",
    "Wrist Pain": "Dolor de muñeca"
}

df_survey["Industry"] = df_survey["Industry"].replace(replace_industry)
df_survey["Gender"] = df_survey["Gender"].replace(replace_gender)
df_survey["Region"] = df_survey["Region"].replace(replace_continent)
df_survey["Job_Role"] = df_survey["Job_Role"].replace(replace_roles)
df_survey["Work_Arrangement"] = df_survey["Work_Arrangement"].replace(replace_work_mode)
df_survey["Mental_Health_Status"] = df_survey["Mental_Health_Status"].replace(replace_mental_health)
df_survey["Burnout_Level"] = df_survey["Burnout_Level"].replace(replace_desgaste)
df_survey["Physical_Health_Issues"] = (df_survey["Physical_Health_Issues"].replace(replace_physical_health, regex=True))
unicos_lista = (df_survey["Physical_Health_Issues"].str.split(";").explode().str.strip().drop_duplicates().tolist())
unicos_lista
```

```python
columna_es = {
    "Survey_Date": "Fecha_encuesta",
    "Age": "Edad",
    "Gender": "Genero",
    "Region": "Region",
    "Industry": "Industria",
    "Job_Role": "Cargo",
    "Work_Arrangement": "Modalidad_de_trabajo",
    "Hours_Per_Week": "Horas_por_semana",
    "Mental_Health_Status": "Estado_salud_mental",
    "Burnout_Level": "Nivel_de_burnout",
    "Work_Life_Balance_Score": "Puntaje_equilibrio_vida_trabajo",
    "Physical_Health_Issues": "Problemas_salud_fisica",
    "Social_Isolation_Score": "Puntaje_aislamiento_social",
    "Salary_Range": "Rango_salarial",
}
df_survey = df_survey.rename(columns={k: v for k, v in columna_es.items() if k in df_survey.columns})
```

```python
df_survey.drop(['Fecha_encuesta', 'Region', "Problemas_salud_fisica", "Rango_salarial"], axis=1, inplace=True)
```

```python
df_survey = df_survey.dropna(subset=["Modalidad_de_trabajo", "Horas_por_semana"])
```

```python
df_survey
```

```python
# asegurar numérico
for col in ["Puntaje_aislamiento_social", "Puntaje_equilibrio_vida_trabajo", "Horas_por_semana"]:
    df_survey[col] = pd.to_numeric(df_survey[col], errors="coerce")

# --- 1. Medias ---
tabla_medias = (
    df_survey
      .groupby("Modalidad_de_trabajo")[["Puntaje_aislamiento_social",
                                        "Puntaje_equilibrio_vida_trabajo",
                                        "Horas_por_semana"]]
      .mean()
      .rename(columns={
          "Puntaje_aislamiento_social": "Aislamiento_media",
          "Puntaje_equilibrio_vida_trabajo": "Equilibrio_vida_media",
          "Horas_por_semana": "Horas_semana_media"
      })
      .round(2)
)

# --- 2. Burnout en % ---
tabla_burnout = (
    df_survey
      .groupby(["Modalidad_de_trabajo", "Nivel_de_burnout"])
      .size()
      .unstack(fill_value=0)
)

tabla_burnout = tabla_burnout[["Bajo", "Medio", "Alto"]]  # orden columnas
tabla_burnout_pct = (tabla_burnout.div(tabla_burnout.sum(axis=1), axis=0) * 100).round(1)

# --- 3. Unir en un solo DataFrame con MultiIndex en columnas ---
final = pd.concat(
    [tabla_medias, tabla_burnout_pct],
    axis=1,
    keys=["Medias", "Burnout (%)"]
)

final
```

```python
# extraer solo las columnas de burnout
tabla_burnout_pct = final["Burnout (%)"]

plt.figure(figsize=(8,5))
sns.heatmap(
    tabla_burnout_pct,
    annot=True, fmt=".1f", cmap="YlOrRd",
    cbar_kws={'label': 'Porcentaje (%)'}
)

plt.title("Distribución de Burnout por Modalidad de Trabajo", fontsize=16, pad=20)
plt.xlabel("Nivel de Burnout", fontsize=12)
plt.ylabel("Modalidad de Trabajo", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
tabla_estres_pct = final["Burnout (%)"]
fig, axes = plt.subplots(1, 3, figsize=(15, 8), sharey=True)
niveles_estres = ['Alto', 'Medio', 'Bajo']
colores = sns.color_palette("colorblind", 3)
for i, (nivel, color_actual) in enumerate(zip(niveles_estres, colores)):
    ax = axes[i]
    datos_nivel = tabla_estres_pct[nivel]

    barras = ax.bar(datos_nivel.index, datos_nivel.values,
                    color=color_actual, width=0.6)

    ax.set_title(f'Estrés {nivel}', fontsize=15, pad=10, fontweight='bold')
    ax.set_ylabel('Porcentaje de Personas (%)', fontsize=13 if i == 0 else 0, labelpad=10)
    ax.set_ylim(0, 100)
    ax.tick_params(axis='x', rotation=25, labelsize=13, pad=5)
    for b in barras:
        h = b.get_height()
        ax.text(b.get_x() + b.get_width()/2,
                h + 1.5, f'{h:.1f}%', ha='center', va='bottom', fontsize=11)

plt.suptitle('Distribución de Burnout por Modalidad de Trabajo', fontsize=18)
plt.tight_layout(pad=2, w_pad=6)
plt.show()
```
