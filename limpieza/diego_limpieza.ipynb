```python
mapa_productividad = {"Low": 0, "Medium": 1, "High": 2}
data["productivity_label_num"] = data["productivity_label"].map(mapa_productividad)

# Calcular la media numérica
media = data["productivity_label_num"].mean()

# Crear el gráfico
plt.figure(figsize=(6,4))
plt.bar(["Media de Productividad"], [media], color="#0077ff")
plt.text(0, media + 0.05, f"{media:.2f}", ha='center', fontsize=12, fontweight='bold')

# Etiquetas y estilo
plt.title("Promedio de Nivel de Productividad", fontsize=14, fontweight='bold')
plt.ylabel("Nivel (0=Bajo, 2=Alto)")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.ylim(0, 2)

plt.show()
```
